```python
from sympy import *
init_printing()          

x, y, z = symbols('x y z')

%matplotlib inline


plot( sqrt( 2*x + 2 ) - 1, 1/2*(x+1)**2 - 1, x, xlim = (-6,6), ylim = (-6,6) )

solve( Eq( sqrt( 2*x + 2 ) - 1 ), x )

solve( Eq( 1/2*(x+1)**2 - 1 ), x )

solve( Eq( 1/2*(x+1)**2 - 1 ), sqrt( 2*x + 2 ) - 1 )
```
